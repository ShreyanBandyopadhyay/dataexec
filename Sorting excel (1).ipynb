```python
import pandas as pd
from openpyxl import load_workbook

# Load your Excel file
excel_file = r'C:\Users\BandyopadhyayS\Downloads\Book1.xlsx'  # Replace with your file path
sheet_name = 'Sheet1'  # Replace with your sheet name
output_file = 'CHEC PSG SAN2.xlsx'  # Replace with desired output file name

# Load the Excel sheet into a pandas DataFrame
df = pd.read_excel(excel_file, sheet_name=sheet_name)

# Define the maximum percentage difference allowed
max_diff = 0.15

# Function to determine if row should be hidden based on percentage difference
def should_hide_row(row):
    value_A = row['A']
    value_B = row['B']
    if pd.notnull(value_A) and pd.notnull(value_B):
        # Check if both values are numeric and value_A is not zero
        if isinstance(value_A, (int, float)) and isinstance(value_B, (int, float)) and value_A != 0:
            percentage_diff = abs(value_B - value_A) / value_A
            return percentage_diff <= max_diff  # Hide rows within 15% range
    return False  # Return False if any value is NaN, non-numeric, or value_A is zero

# Filter rows to hide based on the condition
rows_to_hide = df.apply(should_hide_row, axis=1)

# Write the filtered DataFrame to a new Excel file
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name=sheet_name, index=False)
    writer.book.save(output_file)
    writer.book.close()

# Open the output Excel file with openpyxl and hide rows
book = load_workbook(output_file)
worksheet = book[sheet_name]

for idx, should_hide in enumerate(rows_to_hide):
    if should_hide:
        worksheet.row_dimensions[idx + 2].hidden = True  # Adjust for 1-based index

# Save the output Excel file
book.save(output_file)

print(f"Filtered and hidden rows saved to '{output_file}'")
```
