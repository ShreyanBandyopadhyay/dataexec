```python
import pandas as pd
import numpy as np

def sort_and_remove_values(row):
    if len(row) <= 1:
        return row
    
    # Extract numeric values and ignore non-numeric values
    numeric_values = [value for value in row if isinstance(value, (int, float))]
    
    if len(numeric_values) <= 1:
        return row
    
    highest = max(numeric_values)
    lowest = min(numeric_values)
    
    # Replace non-numeric values with None, while keeping text values intact
    result_row = []
    for value in row:
        if isinstance(value, (int, float)):
            if value == highest or value == lowest:
                result_row.append(value)
            else:
                result_row.append(None)
        else:
            result_row.append(value)
    
    return result_row

def process_excel(input_file, output_file):
    # Read Excel file into a pandas DataFrame
    df = pd.read_excel(input_file, engine='openpyxl')
    
    # Apply the sorting and removal function row-wise
    df = df.apply(sort_and_remove_values, axis=1, result_type='broadcast')
    
    # Write modified DataFrame back to Excel
    df.to_excel(output_file, index=False, engine='openpyxl')

# Example usage:
input_file = r'C:\Users\BandyopadhyayS\Downloads\Book3.xlsx'  # Replace with your input Excel file path
output_file = 'output.xlsx'  # Replace with desired output Excel file path

process_excel(input_file, output_file)
```

```python
import pandas as pd
import numpy as np

def sort_and_remove_values(row):
    if len(row) <= 1:
        return row
    
    numeric_values = []
    text_values = []
    
    # Separate numeric and text values
    for value in row:
        if isinstance(value, (int, float)):
            numeric_values.append(value)
        elif isinstance(value, str):
            text_values.append(value)
        else:
            text_values.append(None)  # Handle other types (like None/NaN)
    
    if len(numeric_values) <= 1:
        return row
    
    highest = max(numeric_values)
    lowest = min(numeric_values)
    
    # Construct the result row
    result_row = []
    for value in row:
        if isinstance(value, (int, float)):
            if value == highest or value == lowest:
                result_row.append(value)
            else:
                result_row.append(None)
        else:
            result_row.append(value)
    
    return result_row

def process_excel(input_file, output_file):
    # Read Excel file into a pandas DataFrame without automatic type inference
    df = pd.read_excel(input_file, engine='openpyxl', dtype=object)
    
    # Apply the sorting and removal function row-wise
    df = df.apply(sort_and_remove_values, axis=1, result_type='broadcast')
    
    # Write modified DataFrame back to Excel
    df.to_excel(output_file, index=False, engine='openpyxl')

# Example usage:
input_file = r'C:\Users\BandyopadhyayS\Downloads\Book3.xlsx'  # Replace with your input Excel file path
output_file = 'output2.xlsx'  # Replace with desired output Excel file path

process_excel(input_file, output_file)
```
